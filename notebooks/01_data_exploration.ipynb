```python
game_lookup = games[
    ["game_id", "season", "date", "competition_id"]
].rename(
    columns={
        "date": "game_date",
        "competition_id": "game_competition_id",
    }
)
```

```python
merged_appearances = appearances.merge(
    game_lookup,
    on="game_id",
    how="left",
    validate="many_to_one",
    indicator=True,
)
```

```python
(
    merged_appearances["date"]
    == merged_appearances["game_date"]
).value_counts(dropna=False)
```

```python
(
    merged_appearances["competition_id"]
    == merged_appearances["game_competition_id"]
).value_counts(dropna=False)
```

```python
merged_appearances = merged_appearances.drop(
    columns=[
        "game_date",
        "game_competition_id",
        "_merge",
    ]
)
```

```python
pl_appearances = merged_appearances.loc[
    merged_appearances["competition_id"] == "GB1"
].copy()
```

```python
pl_appearances.groupby("season").agg(
    appearance_rows=("appearance_id", "count"),
    unique_players=("player_id", "nunique"),
    unique_clubs=("player_club_id", "nunique"),
)
```

```python
player_season_stats = (
    pl_appearances
    .groupby(["season", "player_id", "player_name"], as_index=False)
    .agg(
        appearances=("appearance_id", "count"),
        minutes_played=("minutes_played", "sum"),
        goals=("goals", "sum"),
        assists=("assists", "sum"),
        yellow_cards=("yellow_cards", "sum"),
        red_cards=("red_cards", "sum"),
        clubs_played_for=("player_club_id", "nunique"),
    )
)
```

```python
player_season_stats.shape
```

```python
player_season_stats.head()
```

```python
pl_games = games.loc[
    games["competition_id"] == "GB1"
].copy()

pl_games["date"] = pd.to_datetime(
    pl_games["date"]
)

valuations["date"] = pd.to_datetime(
    valuations["date"]
)
```

```python
season_end_dates = (
    pl_games
    .groupby("season", as_index=False)
    .agg(season_end_date=("date", "max"))
)

season_end_dates
```

```python
player_seasons = player_season_stats.merge(
    season_end_dates,
    on="season",
    how="left",
    validate="many_to_one",
)
```

```python
valuation_lookup = valuations[
    ["player_id", "date", "market_value_in_eur"]
].rename(
    columns={
        "date": "valuation_date",
    }
)
```

```python
valuation_candidates = player_seasons.merge(
    valuation_lookup,
    on="player_id",
    how="left",
    validate="many_to_many",
)
```

```python
valuation_candidates["days_after_season"] = (
    valuation_candidates["valuation_date"]
    - valuation_candidates["season_end_date"]
).dt.days
```

```python
postseason_candidates = valuation_candidates.loc[
    valuation_candidates["days_after_season"] >= 0
].copy()
```

```python
first_postseason_valuations = (
    postseason_candidates
    .sort_values("valuation_date")
    .drop_duplicates(
        subset=["season", "player_id"],
        keep="first",
    )
)
```

```python
first_postseason_valuations.duplicated(
    ["season", "player_id"]
).sum()
```

```python
first_postseason_valuations[
    "days_after_season"
].describe(
    percentiles=[0.25, 0.5, 0.75, 0.9, 0.95]
)
```

```python
labelled_player_seasons = first_postseason_valuations.loc[
    first_postseason_valuations["days_after_season"] <= 120
].copy()
```

```python
scoring_player_seasons = player_seasons.loc[
    player_seasons["season"] == 2025
].copy()
```

```python
coverage_by_season = (
    player_seasons
    .groupby("season")
    .size()
    .rename("total_players")
    .to_frame()
)

coverage_by_season["labelled_players"] = (
    labelled_player_seasons
    .groupby("season")
    .size()
)

coverage_by_season["coverage_percentage"] = (
    coverage_by_season["labelled_players"]
    / coverage_by_season["total_players"]
    * 100
)

coverage_by_season
```

```python
player_attributes = players[
    [
        "player_id",
        "date_of_birth",
        "position",
        "sub_position",
        "foot",
        "height_in_cm",
        "country_of_citizenship",
    ]
].copy()
```

```python
player_attributes["date_of_birth"] = pd.to_datetime(
    player_attributes["date_of_birth"]
)
```

```python
model_data = labelled_player_seasons.merge(
    player_attributes,
    on="player_id",
    how="left",
    validate="many_to_one",
    indicator=True,
)
```

```python
model_data.shape
```

```python
model_data["_merge"].value_counts()
```

```python
model_data["age_at_season_end"] = (
    (
        model_data["season_end_date"]
        - model_data["date_of_birth"]
    ).dt.days
    / 365.25
)
```
